```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Draw n_samples from uniform distrubution inside (bounds,bounds)
bounds = 1000
n_samples = 50000
samples = [(np.random.randint(0,bounds+1),np.random.randint(0,bounds)) for x in range(n_samples)]
```

```python
# Check sample distance from 0,0
in_quadrant = []
out_quadrant = []

for sample in samples:
    if np.sqrt(sample[0]**2 + sample[1]**2) <= bounds:
        in_quadrant.append(sample)
    else:
        out_quadrant.append(sample)
```

```python
# Approximate Pi
pi_approx = 4 * len(in_quadrant) / n_samples
```

```python
# Make lists of points inside and outside of circle quadrant
x_in = [coord[0] for coord in in_quadrant]
x_out = [coord[0] for coord in out_quadrant]
y_in = [coord[1] for coord in in_quadrant]
y_out = [coord[1] for coord in out_quadrant]

# List of thetas to plot 90 degree arc
theta = np.arange(0, (np.pi / 2), .1)
```

```python
#Plot
plt.scatter(x_in, y_in, c='red', s=1, alpha=0.1)
plt.scatter(x_out, y_out, c='blue', s=1, alpha=0.1)
plt.plot(bounds * np.cos(theta), bounds * np.sin(theta), c='cyan')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Monte Carlo Pi Approximation = {pi_approx}')
plt.axis('equal')  # Ensure the axes are proportional
plt.show
```
